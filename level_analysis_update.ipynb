```python
# Hm dng c 4 phng php  chun ha
from scipy import stats
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def analyze_skewness(x, customers):
    fig, ax = plt.subplots(2, 2, figsize=(10, 10))

    # V biu  phn phi
    sns.histplot(customers[x], kde=True, ax=ax[0, 0])
    ax[0, 0].set_title(f'Original - {x}')

    sns.histplot(np.log1p(customers[x]), kde=True, ax=ax[0, 1])
    ax[0, 1].set_title(f'Log Transformation - {x}')

    sns.histplot(np.sqrt(customers[x]), kde=True, ax=ax[1, 0])
    ax[1, 0].set_title(f'Square Root Transformation - {x}')

    sns.histplot(pd.Series(stats.boxcox(customers[x] + 1)[0]), kde=True, ax=ax[1, 1])
    ax[1, 1].set_title(f'Box-Cox Transformation - {x}')

    plt.tight_layout()
    plt.show()

    # Tnh skewness cho tng phng php
    original_skew = customers[x].skew().round(2)
    log_skew = np.log1p(customers[x]).skew().round(2)
    sqrt_skew = np.sqrt(customers[x]).skew().round(2)
    boxcox_skew = pd.Series(stats.boxcox(customers[x] + 1)[0]).skew().round(2)

    # In kt qu skewness
    print(f"Skewness (Original): {original_skew}")
    print(f"Skewness (Log Transformation): {log_skew}")
    print(f"Skewness (Square Root Transformation): {sqrt_skew}")
    print(f"Skewness (Box-Cox Transformation): {boxcox_skew}")
```

```python
analyze_skewness('drop', df_filtered)
analyze_skewness('arpu', df_filtered)
analyze_skewness('exit_rate', df_filtered)
analyze_skewness('adrw_per_user', df_filtered)
analyze_skewness('booster_per_user', df_filtered)
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
df = df_filtered6.copy()
```

```python
df['target_best'] = (df['cluster'] == 1).astype(int)
```

```python
# Prepare the features (drop and arpu) and target
X = df[['drop', 'arpu']]
y = df['target_best']
```

```python
# Initialize and fit the logistic regression model
model = LogisticRegression()
model.fit(X, y)
```

```python
# Output the coefficients, which are our x and y for the formula
coef_drop = model.coef_[0][0]
coef_arpu = model.coef_[0][1]
intercept1 = model.intercept_[0]
```

```python
print("Logistic Regression Model Coefficients:")
print("Coefficient for drop (x):", coef_drop)
print("Coefficient for arpu (y):", coef_arpu)
print("Intercept:", intercept1)
```

```python
df['grade_target'] = df['cluster'].map({1: 2, 0: 1, 2: 0})  # 1 tt nht = 2 im
```

```python
from sklearn.linear_model import LinearRegression

X = df[['drop', 'arpu']]
y = df['grade_target']

model = LinearRegression()
model.fit(X, y)

x = model.coef_[0]
y = model.coef_[1]
intercept2 = model.intercept_
```

```python
print("Logistic Regression Model Coefficients:")
print("Coefficient for drop (x):", x)
print("Coefficient for arpu (y):", y)   
print("Intercept:", intercept2)
```

```python
df['grade2'] = df['drop']*x + df['arpu']*y + intercept2
df['grade1'] = df['drop']*coef_drop + df['arpu']*coef_arpu + intercept1
```
